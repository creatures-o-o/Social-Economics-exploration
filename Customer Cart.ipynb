#### Function to return the cost

This function:
    1. Reads the relevant csv files
    2. Returns the cost for the stamp based on user inputs

```python
def getprice(weight, country, pmethod):

    #create an empty dictionary to store k,v for country and zones
    destination = {}
    reader = csv.DictReader(open('Countries and Zones.csv'))
    for d in reader:
        destination.setdefault(d['Destination country'],d['Zones'])
    
    #read the relevant csv files, the .set_index is so can use the 'Weight' string values
    economy_letters = pd.read_csv('Economy Letters Price Guide ($).csv')
    economy_letters.set_index('Weight', inplace = True)
    express_letters = pd.read_csv('Express Letter Price Guide ($).csv')
    express_letters.set_index('Weight', inplace = True)
    sales = pd.read_csv('sales_history.csv')
    standard_parcel = pd.read_csv('Standard Parcel Price Guide ($).csv')
    standard_parcel.set_index('Weight', inplace = True)
    express_parcel = pd.read_csv('Express Parcel Price Guide ($).csv')
    express_parcel.set_index('Weight', inplace = True)
    air_parcel = pd.read_csv('Economy Parcel Price Guide_by Air ($).csv')
    air_parcel.set_index('Weight', inplace = True)
    sea_parcel = pd.read_csv('Economy Parcel Price Guide_by Sea ($).csv')
    sea_parcel.set_index('Weight', inplace = True)
    
    #returns the zone for the destination
    zone = destination.get(country)  
    
    #If the user has picked Economy letter do the following
    if pmethod == "Economy Letter":
        #This code is to return the right column as the columns have 'zone 2, 3 and 5'
        if zone == "Zone 1":
            n_zone = "Zone 1"
        elif (zone == 'Zone 2' or zone == 'Zone 3' or zone == 'Zone 5'):
            n_zone = 'Zone 2,3 and 5'
        elif (zone =='Zone 4' or zone == 'Zone 6' or zone == 'Zone 7' or zone == 'Zone 9'):
            n_zone = 'Zone 4, 6, 7 and 9'
        elif zone == 'Zone 8':
            n_zone = "Zone 8"
        
        #checks the weight of the letters to return the cost    
        if weight > 0 and weight <= 50:
            cost = economy_letters.at['Up to 50g',n_zone]
        elif weight > 50 and weight <=250:
            cost = economy_letters.at['Over 50g up to 250g',n_zone]
        elif weight > 250 and weight <=500:
            cost = economy_letters.at['Over 250g up to 500g',n_zone]

        return cost
    #returns the cost of the express letters
    elif pmethod == "Express Letter":       
        cost = express_letters.at['Up to 500g',zone]

        return cost
    #checks the weight of the parcel and method of sending to return the cost 
    elif pmethod == "Standard Parcel":
        if weight > 0 and weight <= 0.5:
            cost = standard_parcel.at['Up to 500g',zone]
        elif weight > 0.5 and weight <= 1:
            cost = standard_parcel.at['Up to 1kg',zone] 
        elif weight > 1 and weight <= 1.5:
            cost = standard_parcel.at['Up to 1.5kg',zone]       
        elif weight >1.5 and weight <= 2:
            cost = standard_parcel.at['Up to 2kg',zone] 
        elif weight > 2 and weight <= 3:
            cost = standard_parcel.at['Up to 3kg',zone]                
        elif weight > 3 and weight <= 5:
            cost = standard_parcel.at['Up to 5kg',zone] 
        elif weight > 5 and weight <= 10:
            cost = standard_parcel.at['Up to 10kg',zone] 
        elif weight >10 and weight <= 15:
            cost = standard_parcel.at['Up to 15kg',zone] 
        elif weight > 15 and weight <= 20:
            cost = standard_parcel.at['Up to 20kg',zone]
        return cost
    
    elif pmethod == "Express Parcel":        
        if weight > 0 and weight <= 0.5:
            cost = express_parcel.at['Up to 500g',zone]      
        elif weight > 0.5 and weight <= 1:
            cost = express_parcel.at['Up to 1kg',zone]       
        elif weight > 1 and weight <= 1.5:
            cost = express_parcel.at['Up to 1.5kg',zone]         
        elif weight >1.5 and weight <= 2:
            cost = express_parcel.at['Up to 2kg',zone] 
        elif weight > 2 and weight <= 3:
            cost = express_parcel.at['Up to 3kg',zone]                 
        elif weight > 3 and weight <= 5:
            cost = express_parcel.at['Up to 5kg',zone] 
        elif weight > 5 and weight <= 10:
            cost = express_parcel.at['Up to 10kg',zone] 
        elif weight >10 and weight <= 15:
            cost = express_parcel.at['Up to 15kg',zone] 
        elif weight > 15 and weight <= 20:
            cost = express_parcel.at['Up to 20kg',zone] 
        return cost 
    elif pmethod == "Economy Parcel_By Air":
        if weight > 0 and weight <= 0.5:
            cost = air_parcel.at['Up to 500g',zone]
        elif weight > 0.5 and weight <= 1:
            cost = air_parcel.at['Up to 1kg',zone]      
        elif weight > 1 and weight <= 1.5:
            cost = air_parcel.at['Up to 1.5kg',zone]      
        elif weight >1.5 and weight <= 2:
            cost = air_parcel.at['Up to 2kg',zone]
 
        return cost
    elif pmethod == "Economy Parcel_By Sea":
        if weight > 0 and weight <= 3:
            cost = sea_parcel.at['Up to 3kg',zone]               
        elif weight > 3 and weight <= 5:
            cost = sea_parcel.at['Up to 5kg',zone]
        elif weight > 5 and weight <= 10:
            cost = sea_parcel.at['Up to 10kg',zone]
        elif weight >10 and weight <= 15:
            cost = sea_parcel.at['Up to 15kg',zone]
        elif weight > 15 and weight <= 20:
            cost = sea_parcel.at['Up to 20kg',zone]
    
        return cost
```

#### View cart item
Allows user to view the cart and decide what they want to do

```python
def view_cart(orders):
    #creates a new frame from the cart[{}]
    new = pd.DataFrame(orders)
    #creates an order columns list
    column_titles = ["sale_id","postage_method","destination","quantity","weight", "cost", "total_cost"]
    #renumber the index so that it starts at 1
    new.index = np.arange(1, len(new)+1)
    #rearrage the columns to order of the columns list above
    cart_view = new.reindex(columns = column_titles)
    #use display instead of print as it makes for a nicer table
    display(cart_view) 
    total = sum(item["total_cost"] for item in orders)
    print ("You current cart total is $" + "%.2f" % total)
```
